## Notebooks

Un notebook es un entorno de programación interactivo que te permite escribir y ejecutar código, así como documentar y visualizar tus resultados en un mismo lugar. Es similar a un editor de texto, pero tiene algunas características adicionales que lo hacen ideal para la ciencia de datos y el análisis estadístico.

En un notebook, puedes escribir código en diferentes celdas, lo que te permite dividir tu trabajo en secciones lógicas y ejecutar cada celda de forma independiente. Esto es muy útil para explorar y analizar datos de manera iterativa, ya que puedes ejecutar una celda y luego ajustarla o modificarla antes de ejecutar la siguiente.

Además, un notebook te permite combinar código con texto, imágenes, gráficos y otros elementos multimedia, lo que facilita la documentación y presentación de tus resultados. Puedes agregar títulos, subtítulos y comentarios para explicar tu proceso y tus hallazgos, y puedes insertar visualizaciones para ilustrar tus conclusiones.

Un notebook también puede ser compartido con otros usuarios, lo que lo convierte en una herramienta útil para la colaboración y el trabajo en equipo. Puedes compartir tu notebook con colegas o con la comunidad en línea para obtener retroalimentación y comentarios sobre tu trabajo, o para reproducir y extender los análisis de otros.

En resumen, un notebook es una herramienta esencial para cualquier persona que trabaje con datos y análisis estadístico. Te permite combinar código, texto y visualizaciones en un solo lugar, lo que hace que el proceso de análisis sea más eficiente, iterativo y colaborativo.

## Crear datos ficticios

Otra opción es generar un conjunto de datos simulados para representar la rotación del personal en una empresa. Supongamos que nuestra empresa tiene 100 empleados y queremos simular la rotación en un período de un año. Para ello, podemos utilizar el siguiente código:



```r
set.seed(123) #fija nuestro punto de partida igual para simular datos y obtener siempre los mismos resultados.

# Generamos los datos de empleados, de maanera ficticia. Estas funciones permiten crear datos distribuidos uniforme, entre ciertos valores.

employee_data <- data.frame(
  employee_id = 1:100,
  gender = sample(c("Male", "Female"), 100, replace = TRUE),
  age = sample(18:60, 100, replace = TRUE),
  department = sample(c("Sales", "Marketing", "Finance", "IT"), 100, replace = TRUE),
  salary = sample(30000:80000, 100, replace = TRUE)
)

# Generamos los datos de rotación
turnover_data <- data.frame(
  employee_id = sample(1:100, 20, replace = FALSE),
  termination_date = sample(seq(as.Date('2023/01/01'), as.Date('2023/12/31'), by="day"), 20),
  reason = sample(c("Resignation", "Termination", "Layoff"), 20, replace = TRUE)
)
```



Tenemos dos fuentes de datos, uno con los datos de los empleados "employee_data" y otro con la rotación "turnover_data":



```r
# Unimos los datos
full_data <- merge(employee_data, turnover_data, by = "employee_id", all.x = TRUE)

head(full_data)
```


Podemos ver que hay varios datos llamados <NA>	o NA que "Not Available" (No Disponible). Un dato NA es utilizado para representar valores faltantes o desconocidos en un conjunto de datos. Por ejemplo, si se está trabajando con una base de datos que tiene información sobre la edad de los clientes, es posible que algunos registros no tengan información sobre la edad, por lo que se utiliza el valor NA para indicar que ese valor no está disponible.

En R, el valor NA es utilizado en diferentes funciones para manejar valores faltantes. Por ejemplo, la función is.na() se utiliza para determinar si un valor es NA o no. También hay funciones como na.omit() que se utilizan para eliminar los valores NA de un conjunto de datos, y funciones como na.fill() que se utilizan para rellenar los valores NA con valores predeterminados o calculados. El manejo adecuado de los datos NA es importante para garantizar la precisión y confiabilidad de los análisis y modelos que se construyen a partir de los datos.

Podemos ver que estos graficos son bastante básicos, pero el paquete ggplot2 permite una infinita personalización.
Por ejemplo, un boxplot creado con ggplot2:


```r
#create 

 of price, grouped by cut
ggplot(data=full_data, aes(x=departmen, y=salary)) + 
  geom_boxplot(fill="steelblue") of price, grouped by cut
ggplot(data=full_data, aes(x=departmen, y=salary)) + 
  geom_boxplot(fill="steelblue")
```

```r
library(ggplot2)

ggplot(employee_data, aes(x=department, y=salary)) +
  geom_boxplot(fill="steelblue", color="black") +
  labs(x="Departamento", y="Salario", title="Boxplot del salario por departamento")
```

La función ggplot() establece el conjunto de datos y la estética básica, mientras que la función geom_boxplot() crea el gráfico de caja y bigotes. La opción fill="blue" establece el color de relleno del boxplot en azul, mientras que color="black" establece el color de los bordes en negro. La función labs() se utiliza para establecer los nombres de los ejes y el título del gráfico. Podriamos hacer otra versión con los puntos de los datos distribuidos, para entnder mejor y así, ir personalizando.

```r
ggplot(employee_data, aes(x = department, y = salary)) +
  geom_boxplot(fill = "lightblue", color = "blue") +
  geom_jitter(aes(color = department), width = 0.2, alpha = 0.5) +
  labs(title = "Boxplot del salario por departamento", x = "Departmento", y = "Salario") +
  theme_classic()
```

En este código, la función geom_boxplot() dibuja el boxplot, y la función geom_jitter() agrega puntos detrás del gráfico. El argumento width controla la separación horizontal de los puntos, mientras que alpha controla su transparencia. El argumento color en aes() asigna diferentes colores a cada departamento, y la función labs() define los títulos de los ejes x e y y el título del gráfico. El tema theme_classic() establece el estilo clásico de ggplot2.
